# Limpieza de datos históricos

```python
df_data_historica[df_data_historica['Local'].str.contains('Sweden') & df_data_historica['Visitante'].str.contains('Austria')].index
```

```python
df_data_historica.drop(index = 37, inplace=True)
```

```python
df_data_historica[df_data_historica['Resultado'].str.contains('[^\d-]')]
# [] negación
# cualquier carácter(^) que NO sea un dígito(\d) y NO sea un guion(-)
```

```python
df_data_historica[df_data_historica['Resultado'].str.contains(r'[^​\d–]')]
# [] negación
# cualquier carácter(^) que NO sea un dígito(\d) y NO sea un guion(-)
```

```python
df_data_historica['Resultado'] = df_data_historica['Resultado'].str.replace(r'[^​\d–]', '', regex=True)
```

```python
df_data_historica[df_data_historica['Resultado'].str.contains(r'[^​\d–]')]
```

```python
df_data_historica[['Goles_Locales', 'Goles_Visitantes']] = df_data_historica['Resultado'].str.split('–', expand=True)
df_data_historica
```

```python
df_data_historica.drop('Resultado', axis=1, inplace=True)
df_data_historica
```

```python
df_data_historica.dtypes # verificar el tipo de data
df_data_historica = df_data_historica.astype({'Goles_Locales': int, 'Goles_Visitantes': int, 'Año': int})
```

```python
df_data_historica.dtypes
```

```python
df_data_historica['Goles_Totales'] = df_data_historica['Goles_Locales'] + df_data_historica['Goles_Visitantes']
df_data_historica
```
