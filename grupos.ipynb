# Preparación de grupos

```python
import pandas as pd
from string import ascii_uppercase as alfabeto
import pickle
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
todas_tablas = pd.read_html('https://en.wikipedia.org/wiki/2026_FIFA_World_Cup', storage_options=headers)
```

```python
todas_tablas[102]
```

```python
for letra, i in zip(alfabeto, range(25,109, 7)):
    print(letra,i)
```

```python
dict_tablas = {}

for letra, i in zip(alfabeto, range(25,109, 7)):
    df = todas_tablas[i]
    df.rename(columns={df.columns[1]:'Team'}, inplace=True)
    df.pop('Qualification')
    dict_tablas[f'Grupo {letra}'] = df
    print(df)
    print('\n')
```

```python
with open('dict_table.pkl', 'wb') as output:
    pickle.dump(dict_tablas, output)
```
